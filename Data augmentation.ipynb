##  Creating a fucntion from tensorflow for plotting images

```python
def plotImages(images_arr):
    fig,axes = plt.subplots(1,10,figsize=(10,10))
    axes = axes.flatten()
    for img, ax in zip(images_arr,axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show
```

## Augmenting images 

```python
gen = ImageDataGenerator(rotation_range=10,width_shift_range=0.1,
                        height_shift_range=0.1,shear_range=0.15,zoom_range=0.1,
                        channel_shift_range=10,horizontal_flip=True)
```

```python
choosen_image = random.choice(os.listdir("Machine-Learning\cat-vs-dog/train\dog"))
```

```python
image_path = "Machine-Learning\cat-vs-dog/train/dog/"+choosen_image
```

```python
assert os.path.isfile(image_path)
```

```python
image = np.expand_dims(plt.imread(image_path),0)
plt.imshow(image[0])
```

```python
aug_itr = gen.flow(image)
```

```python
aug_images = [next(aug_itr)[0].astype(np.uint8)for i in range(10)]
```

## Ploting images  

```python
plotImages(aug_images)
```
